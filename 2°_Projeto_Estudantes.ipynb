```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df.head()
```

```python
df['média_notas'] = df[['Nota de matemática','Nota de literatura', 'Nota Redação']].mean(axis=1) #Criei uma nova coluna
# [[...]] = Isso significa que uma lista contém a si mesma
df.head()
```

```python
df.shape
```

```python
plt.figure(figsize=(20,7))
sns.set_theme(style="darkgrid")
plt.title('Média das notas por gêneros')
sns.barplot(data=df, x='Gênero', y='média_notas');
```

```python
plt.figure(figsize=(20,7))
sns.set_theme(style="darkgrid")
plt.title('Média das notas por gênero')
sns.boxplot(data=df, x='Gênero', y='média_notas')
```

```python
#plt.figure(figsize=(20,7)) # Não consegui plotar.
#plt.title('Média de notas x Escolaridade')
sns.set_theme(style="darkgrid")
sns.catplot(
    data=df, 
    x = 'Escolaridade', 
    y='média_notas',
    aspect=5, #Dimensiona o gráfico
    # kind="bar" - Coloca dispersão em colunas
    ci=None
    )
```

```python
sns.catplot(
    data=df,
    x='raça',
    y='média_notas',
    aspect=5,
    #kind="bar" - Coloca dispersão em colunas
    ci=None,
)
```
